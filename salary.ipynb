```python
"""
    Salary Predictor using Linear Regression (One Feature)
"""
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

%matplotlib inline
```

```python
# Training Set
df = pd.read_csv(os.path.join('Data', 'Salary.csv'))
# Number of Training Examples
m = df.shape[0]

# Features: Years of Experience
X = df.iloc[:,0].values
# Adding the intercept term X_0
X = np.column_stack((np.ones(m), X))

# Labels: Salary
y = df.iloc[:,-1].values

plt.scatter(X[:, 1], y[:], s=20, c='r', edgecolor='k')
plt.title('Experience (Years) vs Salary')
plt.xlabel('Experience in Years')
plt.ylabel('Salary')
plt.show()
```

```python
def cost(X, y, theta):
  h = np.dot(X, theta)
  J = (1 / (2 * m)) * np.sum(np.square(h - y))
  
  return J

def gradient(X, y):
  theta = np.zeros(X.shape[1])
  alpha = 0.03
  
  learning = []

  for i in range(1500):    
    learning.append([i, cost(X, y, theta)])
    h = np.dot(X, theta)
    theta = theta - (alpha / m) * ((h - y).dot(X))    
        
  return theta, np.array(learning)

# Computing Parameters Theta using Gradient Descent for the Linear Regression Model
theta, learning = gradient(X, y)
```

```python
plt.scatter(X[:, 1], y[:], s=20, c='r', edgecolor='k')
plt.title('Experience (Years) vs Salary')
plt.xlabel('Experience in Years')
plt.ylabel('Salary')

# Line of Best Fit determined by Gradient Descent 
plt.plot(X[:, 1], np.dot(X, theta), '-')
plt.legend(['Training Data', 'Linear Regression'])

plt.show()
```

```python
# Hypothesis Function (Model)
def compute(x, theta):
  return np.round(np.dot(x, theta))

'''
  Predictions using the Linear Regression Model
''' 
# Predicting for YoE between 3 - 20
for i in range(3, 21):
  print(f'Years of Experience: {i}, Predicted Salary: {compute([1, i], theta)}')
```
